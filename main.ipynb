```python
import pandas as pd
```

```python
df = pd.read_csv("mumbai_court_history_v2.csv")
df['business_date'] = pd.to_datetime(df['business_date'], dayfirst=True, errors='coerce')
df['hearing_date'] = pd.to_datetime(df['hearing_date'], dayfirst=True, errors='coerce')
df = df.sort_values(['cnr_number', 'business_date'])
df['days_delayed'] = df.groupby('cnr_number')['business_date'].diff().dt.days
bottlenecks = df.groupby('purpose')['days_delayed'].agg(['mean', 'sum', 'count']).sort_values(by='sum', ascending=False)
print(bottlenecks.head(10))
```
